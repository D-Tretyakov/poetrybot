```python
from russtress import Accent
from pyphonetics import Soundex
from fonetika.soundex import RussianSoundex
import rusyllab
import nltk
import re
```

```python
soundex = RussianSoundex(delete_first_letter=True, code_vowels=True)
accent = Accent()
```

```python
def encode(word):
    sdx = soundex.transform(word)

    accent_letter = accent.put_stress(word).find('\'')
    rel_accent = accent_letter / len(word)
    num_syl = len(rusyllab.split_words(word.split()))
    if rel_accent <= 0.4:
        rel_accent = 0
    elif rel_accent <= 0.67:
        rel_accent = 1
    else:
        rel_accent = 2

    _, pos = nltk.pos_tag([word], lang="rus")[0]
       
    if pos in ("NONLEX"):
        return "-{}{}{}".format(rel_accent, num_syl, sdx)
    return "{}{}{}{}".format(pos[0], rel_accent, num_syl, sdx)
```

```python
p = re.compile(r'[а-яА-ЯёЁ]+')
with open('Lermontov all (raw text).txt', 'r', encoding='utf-8') as f:
    text = p.findall(f.read().lower())
```

```python
sdx_word = {}
ln = len(set(text))
print(ln)
for i, word in enumerate(set(text)):
    if i % 100 == 0:
        print(i / ln)
    sdx_word[encode(word)] = word
```

```python
import json
with open('SDX_WORD.txt', 'w') as f:
    f.write(json.dumps(sdx_word))
```

```python
lines = []
p = re.compile(r'[а-яА-ЯёЁ]+')
with open('Lermontov all (raw text).txt', 'r', encoding='utf-8') as f:
    for i, line in enumerate(f.readlines()):
        if i % 100 == 0:
            print(i)
        reg = p.findall(line.lower())
        if reg:
            lines.append(' '.join(encode(word) for word in reg))
```
